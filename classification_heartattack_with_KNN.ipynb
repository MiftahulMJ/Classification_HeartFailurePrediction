```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Membaca data heart.csv
data = pd.read_csv ('heart.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#menentukan variabel independen
x = data.drop(['HeartDisease'], axis=1)
x.head()
```

```python
#menentukan variabel dependen
y = data['HeartDisease']
y.head()
```

```python
#membagi data traning dan testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Mengidentifikasi fitur kategori
categorical_features = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'] # Replace with actual categorical column names

# Buat ColumnTransformer untuk menerapkan OneHotEncoding ke fitur kategoris dan passthrough untuk fitur numerik
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), x_train.select_dtypes(exclude=['object']).columns),
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)
    ])


# Fit dan transform data training
scaled_x_train = preprocessor.fit_transform(x_train)

# Transform data testing
scaled_x_test = preprocessor.transform(x_test)
```

```python
#membuat model fungsi Knn untuk klasifikasi data
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(scaled_x_train, y_train)
```

```python
#menentukan nilai prediksi
y_pred = knn.predict(scaled_x_test)
y_pred
```

```python
#menampilkan matriks hasil prediksi
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
#mengukur ketetapan hasil prediksi
print(classification_report(y_test, y_pred))
```
